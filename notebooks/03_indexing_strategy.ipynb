# Indexing Strategy

Combine Dune liquidity with observed Tycho indexing speed from local logs.

```python
from pathlib import Path
import plotly.express as px
import polars as pl

processed = Path("../data/processed")
protocols = pl.read_parquet(processed / "protocol_volumes.parquet")
estimates = pl.read_parquet(processed / "tycho_indexing_estimates.parquet")
joined = estimates.join(
    protocols.select("protocol", "volume_30d_usd", "share_30d"), on="protocol", how="left"
)
joined.sort("volume_30d_usd", descending=True)
```

```python
fig = px.scatter(
    joined.to_pandas(),
    x="estimated_hours_from_config_start",
    y="share_30d",
    text="protocol",
    hover_data=["start_block", "observed_blocks_per_minute"],
)
fig.add_vline(x=10)
fig.show()
```
