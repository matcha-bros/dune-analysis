# Liquidity Scoring

Score target pools and inspect protocol/pair coverage from cached Dune exports.

```python
from pathlib import Path
import plotly.express as px
import polars as pl
from tycho_liquidity_analysis.scoring import score_pools

processed = Path("../data/processed")
pools = pl.read_parquet(processed / "top_pools.parquet")
scored = score_pools(pools)
scored.head(25)
```

```python
fig = px.bar(
    scored.head(25).to_pandas(),
    x="token_pair",
    y="relevance_score",
    color="protocol",
    hover_data=["pool_id", "volume_7d_usd", "volume_30d_usd"],
)
fig.show()
```
